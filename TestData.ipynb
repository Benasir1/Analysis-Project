```python
produce_data = data
```

```python
#stripping KES in values column by using Series.map method
produce_data['Values_in_Ksh'] = produce_data['Values_in_Ksh'].map(lambda x: x.strip('KES') if type(x) is str else x).head(5) 
```

```python
produce_data.head
```

```python
import matplotlib.pyplot as plt
```

```python
#add a month colum since data will be plotted in months
produce_data['month'] = pd.DatetimeIndex(produce_data['Date']).month
```

```python
produce_data.head(3)
```

```python
#fomart integer month column to month name formart
import calendar
d = dict(enumerate(calendar.month_abbr))
produce_data['month'] = produce_data['month'].map(d)
```

```python
produce_data.head(3)
```

```python
produce_data.dtypes #checking data type column per column
```

```python
#changing Value column data type float 
produce_data['Values_in_Ksh'] = produce_data['Values_in_Ksh'].astype(float)
```

```python
produce_data.dtypes
```

```python
produce_data.head(1)
```

```python
#transforming date
produce_data['Date'] = pd.to_datetime(produce_data['Date'],infer_datetime_format=True)
```

```python
produce_data.head()
```

```python
produce_data = produce_data.set_index('month')
```

```python
produce_data.head()
```

```python
produce_data = produce_data.reset_index()
```

```python
produce_data.head()
```

```python
produce_data = produce_data.reset_index('Commodity_Type')
```

```python
figure = plt.plot( 'Date', 'Values_in_Ksh', data=produce_data, marker='o', markerfacecolor='blue', markersize=12, color='skyblue')
figure = plt.plot( 'Date', 'Volume_in_Kgs', data=produce_data, marker='', color='olive')
```
